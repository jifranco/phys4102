```python
import numpy as np
import matplotlib.pyplot as plt 
from pylab import *

# Load text
data = np.loadtxt("sunspots.txt")

# Separated the columns into two variables
Time= data[:,0]
Number_Sunspots = data[:,1]

# Each row in both columns are tied to Time and Number of Sunspots
plt.title("Number of Sunspots from January 1749") # Title the graph
plt.xlabel("Months following January 1749") # Labeled x axis
plt.ylabel("Number of Sunspots") # Labeled y-axis
Time,Number_Sunspots = data[:600,0],data[:600,1]
plt.scatter(Time,Number_Sunspots)

Y_avg =[]
r = 5
for i in range(600):
    summation = sum(Number_Sunspots[i-r:i+r]) 
    
    avg= summation/(2*r)
    
    Y_avg.append(avg)
    
# print(Y_avg)

plot(Time,Number_Sunspots,'g.',label='noisy')
plot(Time,Y_avg,label='averaged')
legend()
show()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500, endpoint = False)
f1 = 100
f2 = 105
wave1 = np.sin(2 * np.pi * f1 * t)
wave2 = np.sin(2 * np.pi * f2 * t)
plt.subplot(2, 1, 1)
plt.plot(t, wave1, label='f1=100Hz')
plt.plot(t, wave2, label='f2=105Hz')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Two Sine Waves')
plt.legend()

combined_wave = wave1 + wave2
plt.subplot(2, 1, 2)
plt.plot(t, combined_wave, label='Combined Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Combined Sine Waves')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Load your data
data_photo = np.loadtxt("2D Photoelectric Array")
Frequency2 = data_photo[:,0]
Voltage2 = data_photo[:,1]

# Define the form of the function you want to fit to your data.
def func(x, m, c):
    return m*x + c

# Use curve_fit to fit your data
popt, pcov = curve_fit(func, Frequency2, Voltage2)



# Plot your data
plt.scatter(Frequency2, Voltage2, label='Data')

# Plot the fit
plt.plot(Frequency2, func(Frequency2, *popt), 'r-', label='Fit: y = {:.3}x + {:.2f}'.format(*popt))

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


#the file is loaded and the first whole column and row are skipped
#the 2 and 3 column are now assigned to death and cases
Deaths,Cases = loadtxt("CovidData.csv",int,skiprows=1,usecols=(1,2), delimiter=",", unpack=True)

death=flip(Deaths)
case_pos=flip(Cases)

#days are calculated using the amount of indices there in death (case could be used)
days = range(0,len(death))

plt.figure(figsize= (10,6))

plt.plot(days,cases_pos,"r.")
plt.plot(days,death*100,"k.")
plt.show()
```

```python
ErrHighDeaths= 0.5+np.sqrt(death+0.25)
ErrLowDeaths= -0.5+np.sqrt(death+0.25)

dayMin= 10
dayMax=50
days1=range(dayMin,dayMax)

plt.figure(figsize= (10,6))
plt.errorbar(days1, death[dayMin:dayMax],
             [ErrLowDeaths[dayMin:dayMax],
             ErrHighDeaths[dayMin:dayMax]],
             ls='',marker='o', mfc='red',
             ms=10,mec='red',
             ecolor='red', label="Poisson")
# plt.errorbar(days,death[dayMin:dayMax],
#          ErrDeaths[dayMin:dayMax],
#         ls='',marker='o', mfc='black',ms=6,
#          mec='black',ecolor='black',label="Gaussian")
    
    
    
    
from scipy.optimize import curve_fit

# Create the fit as a function
# x = dependent var
# a b c are the parameters
def myfit(x,a,b,c):
    return a+b*np.exp(c*x)

param0 = (0,100,0.1)
paramfitWithErr1, paramErr1 = curve_fit(myfit,days1,death[dayMin:dayMax],
                               param0,sigma=ErrHighDeaths[dayMin:dayMax])

fitfun1 = myfit(days1,paramfitWithErr1[0],paramfitWithErr1[1],paramfitWithErr1[2])


plot(days1,fitfun1,"k",label="$A+B*exp(C*t)$")
plt.legend()
plt.show()
```
