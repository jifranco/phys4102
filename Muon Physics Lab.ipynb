# 1st Data set 
## taken for ~ 44 hours 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Load data
muonData, time = np.loadtxt("MuonData_2-21-24.data", delimiter=' ', usecols=(0, 1), unpack=True)

# Initialize list for filtered data
decay = []

# Filter out data points below 40000 that are not back-to-back
for i in range(len(muonData)):
    current_val = muonData[i]
    next_val = muonData[i + 1] if i < len(muonData) - 1 else None
    pre_val = muonData[i - 1] if i < len(muonData) - 1 else None

    if current_val < 40000 and (next_val is None or next_val >= 40000) and pre_val >= 40000:
            decay.append(current_val)

# Define the exponential decay function plus background
def exponential_background(t, A, tau, B):
    return A * np.exp(-t / tau) + B

hist, bins = np.histogram(decay, bins=50)

# #Calculate the bin centers
bin_centers = (bins[1:] + bins[:-1]) / 2






errDecay = np.sqrt(hist)      


plt.errorbar(bin_centers,hist,
        errDecay,
       ls='',marker='o', mfc='black',ms=5,mec='black',ecolor='black',label="Decay")
            



p0=(300, 2000, 1)

popt, pcov = curve_fit(exponential_background,bin_centers,hist, p0)

fitfun = exponential_background(bin_centers, *popt)

plt.plot(bin_centers, fitfun, 'r-', label='Fit: A=%5.3f, tau=%5.3f, B=%5.3f' % tuple(popt))
plt.legend()
            
# Plot histogram
plt.hist(decay, bins=50)
plt.yscale('log')
plt.xlabel("Time between Coincidence")
plt.ylabel("Number of Coincidence(log scale)")
plt.title("Filtered Muon Decay Data 2-21-24")
plt.show()


print(np.sqrt(pcov))

print(len(hist))
print(len(bin_centers))
```

## A = 311.4 $\pm$ 3.5
## $\tau$ = 1942.9 ns $\pm$ 34.2 ns
## Background (B) = .94 $\pm$ .65

# 3rd Data Set 
## taken for ~73 hours with same threshold as 2nd set

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Load data
muonData, time = np.loadtxt("MuonData_2-26-24.data", delimiter=' ', usecols=(0, 1), unpack=True)

# Initialize list for filtered data
decay = []

# Filter out data points below 40000 that are not back-to-back
for i in range(len(muonData)):
    current_val = muonData[i]
    next_val = muonData[i + 1] if i < len(muonData) - 1 else None
    pre_val = muonData[i - 1] if i < len(muonData) - 1 else None

    if current_val < 40000 and (next_val is None or next_val >= 40000) and pre_val >= 40000:
            decay.append(current_val)

# Define the exponential decay function plus background
def exponential_background(t, A, tau, B):
    return A * np.exp(-t / tau) + B

hist, bins = np.histogram(decay, bins=50)

# #Calculate the bin centers
bin_centers = (bins[1:] + bins[:-1]) / 2


#Initial Parameters
p0=(500, 2000, 1)



#Gaussian Error
errDecay = np.sqrt(hist)      

#Plotting Gaussian Error Bars onto the Graphj
plt.errorbar(bin_centers,hist,
        errDecay,
       ls='',marker='o', mfc='black',ms=5,mec='black',ecolor='black',label="Decay")
            

    
#Curve fitting
popt, pcov = curve_fit(exponential_background,bin_centers,hist, p0)

#Fitting Function
fitfun = exponential_background(bin_centers, *popt)
#Plotting the fitting function with the labels
plt.plot(bin_centers, fitfun, 'r-', label='Fit: A=%5.3f, tau=%5.3f, B=%5.3f' % tuple(popt))
plt.legend()
            
            
# Plot histogram
plt.hist(decay, bins=50)
plt.yscale('log')
# plt.xscale('log')
plt.xlabel("Time between Coincidence")
plt.ylabel("Number of Coincidence(log scale)")
plt.title("Filtered Muon Decay Data 2-26-24")
plt.show()


print(np.sqrt(pcov))
        
print(len(decay))
```

## A = 516.1 $\pm$ 5.8
## $\tau$ = 2071.5 ns $\pm$ 37.9 ns
